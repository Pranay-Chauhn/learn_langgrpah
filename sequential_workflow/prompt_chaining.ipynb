```python
from langgraph.graph import StateGraph, START, END
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv
from typing import TypedDict
```

```python
load_dotenv()
```

```python
model = ChatOpenAI()
```

```python
class BlogState(TypedDict) :
    title :str
    outline : str
    content :str
    
```

```python
def gen_outline(state : BlogState) -> BlogState :
    title = state['title']
    prompt = f"Write the detailed outline on {title}"

    outline = model.invoke(prompt).content
    state['outline'] = outline
    return state
```

```python
def gen_content(state : BlogState) -> BlogState :
    title = state['title']
    outline = state['outline']
    prompt = f"Write the blog on {title} using the outline"

    content = model.invoke(prompt).content
    state['content'] = content
    return state
```

```python
graph = StateGraph(BlogState)

# node
graph.add_node("gen_outline", gen_outline)
graph.add_node('gen_content', gen_content)

# edges
graph.add_edge(START, 'gen_outline')
graph.add_edge('gen_outline', 'gen_content')
graph.add_edge('gen_content', END)

# compile
workflow = graph.compile()
```

```python
intial_state = {'title' : 'First man on moon.'}

final_state = workflow.invoke(intial_state)['content']
print(final_state)
```
